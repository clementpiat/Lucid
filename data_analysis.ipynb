```python
import os
import json
import matplotlib.pyplot as plt
import numpy as np

window = 30
```

```python
dates = ["19_10_2020"]
hrs_ = []
as_ = []

for date in dates:
    with open(f"data/{date}/hrs.json") as f:
        hrs_.append(json.load(f))
    with open(f"data/{date}/as.json") as f:
        as_.append(json.load(f))
```

```python
index = 0
hr_treshold = 70
acc_treshold = 2
hrs = hrs_[index]
accelerations = as_[index]

avg_hrs = [np.mean(hrs[i-window:i]) for i in range(window, len(hrs))]
accelerations = [np.log(1 + np.std(accelerations[i-window:i])) for i in range(window, len(hrs))]
plt.plot(avg_hrs)
plt.plot([hr_treshold]*len(avg_hrs), 'red')
plt.show()
plt.plot(accelerations)
plt.plot([acc_treshold]*len(accelerations), 'red')
plt.show()
```
